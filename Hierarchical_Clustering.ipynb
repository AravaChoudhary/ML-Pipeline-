```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data = datasets.load_iris()
X = data.data
y = data.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Applying th PCA --> Reducing the Dimension
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(X_scaled)

pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1] , c=y)
plt.title('Scatter Plot')
```

```python
# Hierarchical Clustering --> Agglomerative Clustering
# Making Dendogram
import scipy.cluster.hierarchy as sc

plt.figure(figsize = (20,7))
plt.title('Dendrogram')

sc.dendrogram(sc.linkage(pca_scaled,method = 'ward'))
plt.xlabel('Sample Data')
plt.ylabel('Euclidean Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1] , c=cluster.labels_)
plt.title('Scatter Plot after Clustering and PCA')
```

```python
from sklearn.metrics import silhouette_score

silhoutte_coef = []
for k in range(2,10):
  agglo = AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
  agglo.fit(X_scaled)
  score = silhouette_score(X_scaled,agglo.labels_)
  silhoutte_coef.append(score)

silhoutte_coef

plt.plot(range(2,10),silhoutte_coef)
plt.xticks(range(2,10))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhoutte Coefficient")
plt.title("Silhouette Coefficient Plot")
plt.show()
```
